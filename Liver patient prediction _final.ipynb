### Bilirubin
Bilirubin is a yellow compound that forms in the body as a result of the breakdown of heme, a component of hemoglobin found in red blood cells.Bilirubin is primarily produced in the liver, where it is conjugated (chemically modified) to make it water-soluble. This conjugated bilirubin is then excreted into the bile, which is a digestive fluid produced by the liver and stored in the gallbladder. From there, it enters the intestines and is eventually eliminated from the body in the feces.Elevated levels of bilirubin in the blood can lead to a condition called jaundice, which is characterized by yellowing of the skin and eyes. Jaundice can be caused by various factors, including liver disease, hemolytic anemia (excessive destruction of red blood cells), or bile duct obstruction.
#### Total bilirubin :
It refers to the sum of both direct (conjugated) and indirect (unconjugated) bilirubin levels in the blood. It's a measurement used to assess the overall amount of bilirubin present in the bloodstream.The normal range of total bilirubin levels in the human body typically falls between 0.2 to 1.2 milligrams per deciliter (mg/dL) of blood.
### Direct bilirubin 
Direct bilirubin is the form of bilirubin that has been processed by the liver and is water-soluble. It is conjugated (chemically modified) in the liver to make it easier for the body to excrete.
Indirect bilirubin, is the form of bilirubin that has not yet undergone this liver processing and is less water-soluble.
Total bilirubin is an important marker in diagnosing liver diseases, hemolytic anemias and jaundice. Abnormal levels of total bilirubin can indicate problems with these organs which may require further investigation and medical attention. The reference range for total bilirubin can vary depending on the laboratory and the specific units used for measurement, but typically, normal values fall within a specific range. Your healthcare provider will interpret your total bilirubin levels in the context of your overall health and medical history to make a diagnosis and determine the appropriate course of action.The normal range for direct bilirubin levels in the human body is typically less than 0.3 milligrams per deciliter (mg/dL) of blood.
### ALP(Alkaline phosphatase )
Alkaline phosphatase (ALP) is an enzyme found in various tissues throughout the body, with its highest concentrations typically in the liver, bile ducts, bones, and the lining of the intestines. The liver produces ALP and releases it into the bloodstream. Elevated levels of ALP in the blood can be indicative of liver or biliary (bile duct) issues.the normal range for ALP in the liver is typically between 10 to 40 units per liter (U/L) of blood. 
### ALT(Alanine aminotransferase)
Alanine aminotransferase (ALT), also known as SGPT (serum glutamate pyruvate transaminase), is an enzyme found primarily in the liver.When liver cells are damaged or injured, ALT can leak into the bloodstream, causing elevated levels.Elevated ALT levels can be associated with conditions such as hepatitis, cirrhosis and fatty liver disease.In general, the normal range for ALT in adults is typically considered to be:
For men: 7 to 56 units per liter (U/L) of blood.
For women: 7 to 45 U/L of blood.
### AST(aspartate aminotransferase)
Aspartate aminotransferase (AST), also known as SGOT (serum glutamic oxaloacetic transaminase), is an enzyme that is found in various tissues throughout the body, with higher concentrations in the liver, heart, muscles, and kidneys. AST is released into the bloodstream when these tissues are damaged or injured. 
The normal range for total AST levels in adults is considered to be 10 to 40 units per liter (U/L) of blood for men and women.
### Total_Protiens
It refers to the total concentration of proteins present in the blood plasma. These proteins include various types,such as albumin, globulins, fibrinogen, and other smaller proteins.The normal range for total protein levels falls between 6.0 to 8.3 grams per deciliter (g/dL) for adults.
The two main categories of proteins measured in total protein tests are:
#### Albumin:
This is the most abundant protein in blood plasma and is primarily synthesized by the liver. It plays a crucial role in maintaining the osmotic pressure of the blood and transporting various substances, including hormones, drugs, and certain ions.The normal range for serum albumin levels, which includes albumin produced by the liver, is typically between 3.4 to 5.4 grams per deciliter (g/dL) of blood in adults.
Low albumin levels (hypoalbuminemia) can indicate liver disease, kidney disease, malnutrition, and inflammation, among other conditions. High albumin levels (hyperalbuminemia) are less common and can be associated with dehydration or certain rare genetic disorders.
#### Globulins:
Globulins are a group of proteins that include antibodies (immunoglobulins) and enzymes, among others. They also contribute to the immune system's function and are involved in various physiological processes.

The balance between albumin and globulins in the blood is important for maintaining osmotic pressure, transporting various substances, and supporting immune function. The specific composition of globulins can vary among individuals, and changes in globulin levels can be indicative of various medical conditions, including immune disorders, liver diseases, and certain infections.
### A/G_ratio
It represents the ratio of albumin to globulins in the blood.The A/G ratio provides information about the balance between these two major categories of proteins.The normal A/G ratio is typically greater than 1.0, meaning that there is more albumin than globulins in the blood. This balance is important for maintaining osmotic pressure, transporting substances, and supporting immune function.
##### Interpretation of the A/G ratio can provide insights into various medical conditions:
- If the A/G ratio is lower than the normal range, it may suggest a decrease in albumin levels, which can be seen in conditions such as liver disease, malnutrition, or kidney disorders.
- If the A/G ratio is higher than the normal range, it may indicate an increase in globulin levels, often due to immune system disorders or chronic inflammation.

### Balancing the data using SMOTE

```python
#applying  SMOTE(over sampling technique) to balance the data
from collections import Counter# importing counter to check count of each label
from imblearn.over_sampling import SMOTE 
smote=SMOTE() #object creation
```

```python
print(Counter(y))# checking count for each class 
```

```python
# Smote is always applied on train data(x_train and y_train)
x_smote, y_smote = smote.fit_resample(x_train,y_train)

# checking the balanced data
print("Actual Classes",Counter(y_train))
print("SMOTE Classes",Counter(y_smote))
```

```python
print(x_smote.shape)
print(y_smote.shape)
```

## Model Creation
### Logistic Regression

```python
## Import the Logistic regression model
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_smote,y_smote)# training the model with balanced data
```

```python
y_predict=LR.predict(x_test)
y_predict
```

```python
# Evaluation metrics for classification algorithm
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score,classification_report
```

```python
LR_f1= f1_score(y_test, y_predict)
LR_accuracy=accuracy_score(y_test,y_predict)
LR_precision = precision_score(y_test,y_predict)
LR_recall= recall_score(y_test,y_predict)
LR_CM=confusion_matrix(y_test,y_predict)
print('F1_score:',LR_f1,'\nAccuracy:',LR_accuracy,'\nPrecision:',LR_precision,'\nRecall:',LR_recall,'\nConfusion_matrix:',LR_CM)
```

```python
pd.crosstab(y_test,y_predict)
```

Note: Type 1 error is False positve,Type 2 error is False Negative. Here Type 2 error is more significant than type 1 error because patient having the liver desease should not be predicted as normal which may leads to life threatening case.Hence
We need to reduce type 2 error by increasing the recall score. Hence Recall and F1 score is important for the model.

```python
print(classification_report(y_test, y_predict))
```

```python
## ROC -AUC method
from sklearn.metrics import roc_curve,auc
false_positive_rate,true_positive_rate,thresholds=roc_curve(y_test,y_predict)
roc_auc_score=auc(false_positive_rate, true_positive_rate)
print(roc_auc_score)
```

### Applying KNN Algorithm

```python
# import KNN algorithm
from sklearn.neighbors import KNeighborsClassifier
# create empty list to store error values
error_rate=[]

for i in range(1,11):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(x_smote,y_smote)
    y_predict=model.predict(x_test)
    error_rate.append(np.mean(y_predict!=y_test))
error_rate
```

From the graph we can conclude that k=3 gives the less error

### Conclusion

- The XG Boost algorithm achieved the highest recall score of 96.59%  with f1 score of 88.54%. XG boost algorithm gives better results for imbalanced data as compared to the data which is balanced using smote technique.


- Type 1 error is False Positve and Type 2 error is False Negative. Here Type 2 error is more significant than Type 1 error because patient having the liver desease should not be predicted as normal which may leads to life threatening case. Hence We need to reduce Type 2 error by increasing the recall score. Hence Recall and F1 scores are used to select the XG boost algorithm.


- As per EDA the prominent features that effect the target variable are albumin, age and gender. People having albumin level less than 3.5g/dl can be diagnosed as liver petients as albumin is directly proportional to total proteins and AG ratio.Male patients above the age of 45 have more chance of getting liver disease.
